```python
import math
import pandas as pd
def create_space(lat, lon, s=0.1):
    """Crée une s km par s km surface carré centré sur les coordonnées (latitute,longitude)"""
    lat=float(lat)
    lon=float(lon)
    v = (180/math.pi)*(500/6378137)*s # roughly 0.045 for s=0.25
    return lat - v, lon - v, lat + v, lon + v
```

```python
def add_nightlights(df, tif, tif_array):
    ''' 
    This takes a dataframe with columns cluster_lat, cluster_lon and finds the average 
    nightlights in 2018 using a 0.25kmx0.25km box around the point
    
    I try all the nighlights tifs until a match is found, or none are left upon which an error is raised
    '''
    cluster_nightlights = []
    for i,r in df.iterrows():
        min_lat, min_lon, max_lat, max_lon = create_space(r.GPS__Latitude, r.GPS__Longitude)
        
        xminPixel, ymaxPixel = tif.proj_to_raster(min_lon, min_lat)
        xmaxPixel, yminPixel = tif.proj_to_raster(max_lon, max_lat)
        assert xminPixel < xmaxPixel, print(r.GPS__Latitude, r.GPS__Longitude)
        assert yminPixel < ymaxPixel, print(r.GPS__Latitude, r.GPS__Longitude)
        if xminPixel < 0 or xmaxPixel >= tif_array.shape[1]:
            print(f"no match for {r.GPS__Latitude}, {r.GPS__Longitude}")
            raise ValueError()
        elif yminPixel < 0 or ymaxPixel >= tif_array.shape[0]:
            print(f"no match for {r.GPS__Latitude}, {r.GPS__Longitude}")
            raise ValueError()
        xminPixel, yminPixel, xmaxPixel, ymaxPixel = int(xminPixel), int(yminPixel), int(xmaxPixel), int(ymaxPixel)
        cluster_nightlights.append(np.nanmean(tif_array[yminPixel:ymaxPixel,xminPixel:xmaxPixel]))
        
    df['nightlights'] = cluster_nightlights
```

```python
import pandas as pd
import folium

#fichier_sortie est le nom du fichier où sauvegarder la carte
def creer_carte_gps(dataframe, fichier_sortie):
    # Calcul des moyennes des latitudes et longitudes pour centrer la carte
    moyenne_latitude = dataframe['GPS__Latitude'].mean()
    moyenne_longitude = dataframe['GPS__Longitude'].mean()
    
    # Création de la carte avec folium
    carte = folium.Map(location=[moyenne_latitude, moyenne_longitude], zoom_start=7)
    
    # Ajout des points GPS à la carte
    for _, row in dataframe.iterrows():
        folium.Marker([row['GPS__Latitude'], row['GPS__Longitude']]).add_to(carte)
    
    # Sauvegarde de la carte dans un fichier HTML
    carte.save(fichier_sortie)
    print(f"Carte sauvegardée sous : {fichier_sortie}")
```

create_space est conçue pour créer une surface carrée centrée autour d'un point géographique donné, défini par ses coordonnées de latitude et de longitude. L'objectif principal de cette fonction est de calculer les coordonnées des quatre coins d'un carré autour d'un point central, de manière à ce que chaque côté du carré mesure s kilomètres de long

```python
import math
import pandas as pd
def create_space(lat, lon, s):
    """Crée une surface de s mètres par s mètres de surface carré centré sur les coordonnées (latitute,longitude)"""
    lat=float(lat)
    lon=float(lon)
    v1 = (180/math.pi)*(s/6378137)
    v2=  (180/math.pi)*(s/6378137)/math.cos(lat)
    return lat - v1, lon - v2, lat + v1, lon + v2
```

```python
def visualize_on_map(lat,lon, min_lat, min_lon, max_lat, max_lon):
    
    
    # Crée un Map
    m = folium.Map(location=[lat,lon], zoom_start=16)

    # Mettre des marqueurs au centre
    folium.Marker([lat,lon], popup='Center', icon=folium.Icon(color='red')).add_to(m)

    # Mettre des marqueurs pour les coins
    folium.Marker([min_lat, min_lon], popup='Min Lat, Min Lon', icon=folium.Icon(color='green')).add_to(m)
    folium.Marker([max_lat, max_lon], popup='Max Lat, Max Lon', icon=folium.Icon(color='blue')).add_to(m)
    
    # Dessine le rectangle
    bounds = [(min_lat, min_lon), (max_lat, max_lon)]
    folium.Rectangle(bounds=bounds, color='#ff7800', fill=True, fill_opacity=0.2).add_to(m)

    # Montre le Map
    return m
```

### Les intensités lumineuses nocturnes considérant un rayon de 100 mètres

```python
import math
import rasterio
from rasterio.mask import mask
from shapely.geometry import box
import numpy as np
import pandas as pd

path_to_tiff=r'C:\Users\d\Downloads\VNL\VNL_v21_npp_2018_global_vcmslcfg_c202205302300.average.dat.tif'

def create_space_adjusted(lat, lon, s):
    """
    Calcul l'ajustement en degrés pour une distance `s` en mètres autour d'un point donné (lat, lon),
    prenant en compte la diminution de la distance couverte par un degré de longitude avec l'augmentation de la latitude.
    """
    # Rayon moyen de la Terre en mètres
    earth_radius = 6378137
    # Distance correspondant à un degré à l'équateur
    distance_per_degree = (2 * math.pi * earth_radius) / 360
    # Ajustement pour `s` mètres en degrés
    adjustment = s / distance_per_degree
    # Ajustement pour la longitude
    adjustment_lon = adjustment / math.cos(math.radians(lat))
    return box(lon - adjustment_lon, lat - adjustment, lon + adjustment_lon, lat + adjustment)

def calculate_nightlights_custom(df, path_to_tiff, distance_m=100):
    """
    Rempli un DataFrame avec l'intensité moyenne des lumières nocturnes pour des points définis par GPS__Latitude et GPS__Longitude,
    en utilisant une boîte de taille spécifiée autour de chaque point pour le calcul.
    """
    with rasterio.open(path_to_tiff) as src:
        nightlights_100m = []
        for index, row in df.iterrows():
            # Créer une géométrie de boîte autour du point avec un ajustement pour `distance_m` mètres
            geom = create_space_adjusted(row['GPS__Latitude'], row['GPS__Longitude'], distance_m / 2)  # /2 pour obtenir la moitié de la distance de chaque côté
            
            # Utiliser mask pour découper l'image raster selon la géométrie calculée
            out_image, out_transform = mask(src, [geom], crop=True, all_touched=True)
            
            # Traitement des valeurs no_data du raster
            out_image = out_image.astype(float)
            out_image[out_image == src.nodata] = np.nan  # Convertir les no_data en NaN
            
            # Calcul de la moyenne, en excluant les NaN
            mean_val = np.nanmean(out_image)
            nightlights_100m.append(mean_val)
        
        df['nightlights_100m'] = nightlights_100m

    return df
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import Point, box
import math

def create_space_adjusted(lat, lon, s):
    earth_radius = 6378137
    distance_per_degree = (2 * math.pi * earth_radius) / 360
    adjustment = s / distance_per_degree
    adjustment_lon = adjustment / math.cos(math.radians(lat))
    return box(lon - adjustment_lon, lat - adjustment, lon + adjustment_lon, lat + adjustment)

def visualize_box_and_save(lat, lon, distance_m=100, save_path='C:/Users/d/Desktop/Programme du projet sur CNN pauvreté3/graphs/box_visualization_100m.png'):
    geom = create_space_adjusted(lat, lon, distance_m / 2)
    
    gdf_box = gpd.GeoDataFrame({'id': [1], 'geometry': [geom]})
    gdf_point = gpd.GeoDataFrame({'id': [1], 'geometry': [Point(lon, lat)]})
    
    fig, ax = plt.subplots()
    gdf_box.boundary.plot(ax=ax, color='red', label='100m Box')
    gdf_point.plot(ax=ax, color='blue', marker='o', label='Point')
    
    plt.legend()
    plt.xlabel('Longitude')
    plt.ylabel('Latitude')
    plt.title('Visualisation de boite de 100m autour dun Point')

    # Sauvegarder l'image
    plt.savefig(save_path, bbox_inches='tight')
    plt.close()
```

#### Automatisation

```python
import math
import rasterio
from rasterio.mask import mask
from shapely.geometry import box
import numpy as np
import pandas as pd

def create_space_adjusted(lat, lon, s):
    earth_radius = 6378137  # Rayon moyen de la Terre en mètres
    distance_per_degree = (2 * math.pi * earth_radius) / 360  # Distance par degré à l'équateur
    adjustment = s / distance_per_degree  # Ajustement en degrés pour s mètres
    adjustment_lon = adjustment / math.cos(math.radians(lat))  # Ajustement longitude
    return box(lon - adjustment_lon, lat - adjustment, lon + adjustment_lon, lat + adjustment)

def calculate_nightlights_custom(df, path_to_tiff, distance_m):
    column_name = f'nightlights_{distance_m}m'
    with rasterio.open(path_to_tiff) as src:
        nightlights = []
        for index, row in df.iterrows():
            geom = create_space_adjusted(row['GPS__Latitude'], row['GPS__Longitude'], distance_m / 2)
            out_image, out_transform = mask(src, [geom], crop=True, all_touched=True)
            out_image = out_image.astype(float)
            out_image[out_image == src.nodata] = np.nan  # Convertir les no_data en NaN
            mean_val = np.nanmean(out_image)  # Calcul de la moyenne, en excluant les NaN
            nightlights.append(mean_val)
        df[column_name] = nightlights
    return df

def automate_process(df, path_to_tiff, distance_m):
    base_path = 'C:/Users/d/Desktop/Programme du projet sur CNN pauvreté3/graphs/'
    
    # Calculer l'intensité lumineuse pour chaque point avec la distance spécifiée
    df = calculate_nightlights_custom(df, path_to_tiff, distance_m)
    
    # Génération d'un histogramme global pour la distance_m spécifiée
    histogram_path = f'{base_path}histogramme_nightlights_{distance_m}m.png'
    plt.hist(df[f'nightlights_{distance_m}m'].dropna(), bins=20, edgecolor='black')
    plt.title(f'Histogramme des Lumières Nocturnes {distance_m}m')
    plt.xlabel('Intensité Lumineuse Nocturne')
    plt.ylabel('Fréquence')
    plt.savefig(histogram_path)
    plt.close()
    
    # Afficher des statistiques résumées et le nombre de valeurs NaN pour la nouvelle colonne
    nan_count = df[f'nightlights_{distance_m}m'].isna().sum()
    print(f"Nombre de valeurs NaN dans 'nightlights_{distance_m}m': {nan_count}")
    
    summary_stats = df[f'nightlights_{distance_m}m'].describe()
    print(summary_stats)
```

```python
# Étape 1 : Importation des bibliothèques nécessaires
import pandas as pd
from sklearn.mixture import GaussianMixture
import numpy as np
import matplotlib.pyplot as plt

# Étape 2 : Chargement des données
file_path = 'C:/Users/d/Desktop/Programme du projet sur CNN pauvreté3/DataCIV2.csv'  # Chemin du fichier à ajuster selon le besoin
data = pd.read_csv(file_path)

# Étape 3 : Sélection de la colonne 'nightlights_100m'
X = data['nightlights_100m'].values.reshape(-1, 1)  # Redimensionnement nécessaire pour le GMM

# Étape 4 : Application du GMM pour trouver le nombre optimal de clusters
n_components_range = range(1, 8)  # Test de 1 à 7 clusters
bics = []
for n_components in n_components_range:
    gmm = GaussianMixture(n_components=n_components, random_state=42)
    gmm.fit(X)
    bics.append(gmm.bic(X))
# Trouver le nombre optimal de composants en fonction du score BIC le plus bas
optimal_n_components = n_components_range[np.argmin(bics)]

# Étape 5 : Ajustement du modèle GMM avec le nombre optimal de composantes
gmm_optimal = GaussianMixture(n_components=optimal_n_components, random_state=42)
gmm_optimal.fit(X)

# Étape 6 : Assignation de chaque point de données à son cluster
clusters = gmm_optimal.predict(X)

# Étape 7 : Ajout des assignations de cluster au dataframe
data['cluster'] = clusters

# Affichage des premières lignes pour vérifier les résultats
print(data.head())
print(gmm)
print(gmm_optimal.fit(X))
print(gmm_optimal)
print(clusters)
```

```python
# Examination du modèle GMM

# Nombre de clusters
print(f"Nombre optimal de clusters (composants) : {optimal_n_components}")

# Probabilités d'appartenance aux clusters pour quelques points
probabilities = gmm_optimal.predict_proba(X)
print("Probabilités d'appartenance aux clusters pour les 5 premiers points :")
print(probabilities[:5])

# Moyennes des composants
print("Moyennes des clusters :")
print(gmm_optimal.means_)

# Covariances des composants
print("Covariances des clusters :")
print(gmm_optimal.covariances_)

# Poids des composants (la probabilité a priori de chaque composant)
print("Poids des clusters :")
print(gmm_optimal.weights_)

# Visualisation des moyennes des clusters
plt.figure(figsize=(10, 6))
plt.title("Moyennes des clusters")
plt.scatter(range(optimal_n_components), gmm_optimal.means_.reshape(-1), color='red')
plt.xlabel("Cluster")
plt.ylabel("Moyenne de nightlights_100m")
plt.xticks(range(optimal_n_components))
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.mixture import GaussianMixture
import matplotlib.pyplot as plt

# Chargement des données
data = pd.read_csv('C:/Users/d/Desktop/Programme du projet sur CNN pauvreté3/Data/CIV/Test_Nightlights.csv')
X = data[['nightlights_100m']].values

# Plage de nombres de clusters à tester
n_components_range = range(1, 25)

# Initialisation des listes pour enregistrer les scores
bic_scores = []
aic_scores = []

# Calcul de AIC et BIC pour chaque nombre de clusters
for n_components in n_components_range:
    gmm = GaussianMixture(n_components=n_components, random_state=42).fit(X)
    bic_scores.append(gmm.bic(X))
    aic_scores.append(gmm.aic(X))

# Trouver le nombre de clusters qui minimise le BIC et l'AIC
optimal_clusters_bic = n_components_range[np.argmin(bic_scores)]
optimal_clusters_aic = n_components_range[np.argmin(aic_scores)]

print(f"Nombre optimal de clusters selon BIC: {optimal_clusters_bic}")
print(f"Nombre optimal de clusters selon AIC: {optimal_clusters_aic}")

# Génération et sauvegarde des graphiques
plt.figure(figsize=(8, 6))
plt.plot(n_components_range, bic_scores, '-o', label='BIC')
plt.title('BIC pour différents nombres de clusters')
plt.xlabel('Nombre de clusters')
plt.ylabel('BIC Score')
plt.legend()
plt.tight_layout()
plt.savefig('C:/Users/d/Desktop/Programme du projet sur CNN pauvreté3/Data/CIV/BIC_Score.png')

plt.figure(figsize=(8, 6))
plt.plot(n_components_range, aic_scores, '-o', label='AIC')
plt.title('AIC pour différents nombres de clusters')
plt.xlabel('Nombre de clusters')
plt.ylabel('AIC Score')
plt.legend()
plt.tight_layout()
plt.savefig('C:/Users/d/Desktop/Programme du projet sur CNN pauvreté3/Data/CIV/AIC_Score.png')
```
